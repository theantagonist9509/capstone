```python
import pandas as pd

dfs = [
    pd.read_csv("../dataset/ISIC_2018/ISIC2018_Task3_Training_GroundTruth.csv"),
    pd.read_csv("../dataset/ISIC_2018/ISIC2018_Task3_Validation_GroundTruth.csv"),
    pd.read_csv("../dataset/ISIC_2018/ISIC2018_Task3_Test_GroundTruth.csv"),
]

for df_2018 in dfs:
    # --- Label analysis for df_2018 ---
    label_cols = df_2018.select_dtypes(include='float64').columns.tolist()
    print('Float64 label columns:', label_cols)

    # Check if one-hot encoded:
    # Conditions: all values in {0.0, 1.0}, and each row sums to exactly 1
    vals_ok  = df_2018[label_cols].isin([0.0, 1.0]).all().all()
    rows_sum = df_2018[label_cols].sum(axis=1)
    sums_ok  = (rows_sum == 1.0).all()
    is_one_hot = vals_ok and sums_ok
    print(f'All values are 0 or 1 : {vals_ok}')
    print(f'Every row sums to 1   : {sums_ok}')
    print(f'=> Labels are one-hot  : {is_one_hot}')

    # Column sums (i.e. count per class)
    print('\nColumn sums (samples per class):')
    print(df_2018[label_cols].sum().to_string())

    # Malignant vs Benign counts
    malignant_cols = ['MEL', 'BCC', 'AKIEC']
    benign_cols    = [c for c in label_cols if c not in malignant_cols]
    n_malignant = int(df_2018[malignant_cols].sum(axis=1).clip(upper=1).sum())
    n_benign    = int(df_2018[benign_cols].sum(axis=1).clip(upper=1).sum())
    print(f'\nMalignant (MEL + BCC + AKIEC) : {n_malignant}')
    print(f'Benign (others)               : {n_benign}')
    print(f'Total                         : {n_malignant + n_benign}')
```
